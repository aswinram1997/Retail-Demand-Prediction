## 1. Introduction

In this project, the multivariate LSTM algorithm is being utilized for demand prediction using Walmart's dataset. The LSTM (Long Short-Term Memory) model is a type of recurrent neural network (RNN) that effectively captures long-term dependencies in sequential data. In the context of demand forecasting, the LSTM model is being trained to learn the patterns and relationships between the historical sales data and the corresponding external features such as temperature, fuel price, CPI, unemployment, and holiday events. By incorporating these additional factors, the multivariate LSTM aims to capture the potential impact of external factors on sales patterns and make more accurate predictions. The LSTM architecture is being explored for both sequence-to-one and sequence-to-sequence models, considering different network configurations, layer structures, activation functions, and hyperparameter settings to find the best-performing model. Through the step-by-step implementation of the multivariate LSTM model using the Walmart dataset, this notebook aims to showcase the capabilities of LSTM and its application for accurate sales forecasting within the context of Walmart.

## 2. Methodology

The methodology for the multivariate LSTM in this project involves several key stages. The dataset, consisting of historical sales data along with additional features such as temperature, fuel price, CPI, unemployment, and holiday events, is collected. Data preprocessing tasks are conducted, including data cleaning, feature engineering, and feature scaling. The dataset is then split into training, validation, and testing subsets. The multivariate LSTM model is selected and implemented, exploring different network configurations, layer structures, activation functions, and hyperparameters. The model is trained on the training set and evaluated using the validation set, utilizing performance metrics such as mean squared error (MSE) and root mean squared error (RMSE). Hyperparameter tuning is performed based on the evaluation results to improve the model's performance. Finally, the trained multivariate LSTM model is used to make predictions on the testing set. This methodology ensures robust data preprocessing, appropriate dataset splitting, and rigorous modeling and evaluation, providing reliable insights for decision-making and optimization in the retail industry.

## 3. Development of Demand Prediction Model

### 3.1 Import Libraries

```python
# data manipulation
import numpy as np
import pandas as pd
import random

# data visualization
import matplotlib.pyplot as plt

# data preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# modeling
import os
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, TimeDistributed
from tensorflow.keras import regularizers
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

### 3.2 Data Collection

This phase involves the collection of multivariate data from a CSV file for the Walmart sales dataset and stores it in a dataframe

```python
# Data Reading
df = pd.read_csv('data/Walmart.csv')

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# rearrange columns
df= df[['Store', 'Date', 'Holiday_Flag', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment', 'Weekly_Sales']]

df.head()
```

### 3.3 Data Preprocessing - Feature Engineering

In this phase, new features are created by grouping the data by date. It aggregates the data and calculates various statistics for features such as temperature, fuel price, CPI, unemployment, and weekly sales. By computing metrics like mean, median, maximum, minimum, standard deviation, and variance, new columns are generated in the dataset representing these statistics.

```python
# Group the data by date, considering the common holiday flag
data = df.groupby('Date', as_index=False).agg({
    'Holiday_Flag': 'first',  # Since holiday flag is the same for each date
    'Temperature': ['mean', 'median', 'max', 'min', 'std', 'var'], # Reflects average for all stores
    'Fuel_Price': ['mean', 'median', 'max', 'min', 'std', 'var'], # Reflects average for all stores
    'CPI': ['mean', 'median', 'max', 'min', 'std', 'var'], # Reflects average for all stores
    'Unemployment': ['mean', 'median', 'max', 'min', 'std', 'var'], # Reflects average for all stores
    'Weekly_Sales': 'sum' # Sums the sales for all stores
})

# column names
data.columns = ['Date', 'Holiday_Flag',
           'Temperature_Mean', 'Temperature_Median', 'Temperature_Max', 'Temperature_Min', 'Temperature_Std', 'Temperature_Var',
           'Fuel_Price_Mean', 'Fuel_Price_Median', 'Fuel_Price_Max', 'Fuel_Price_Min', 'Fuel_Price_Std', 'Fuel_Price_Var',
           'CPI_Mean', 'CPI_Median', 'CPI_Max', 'CPI_Min', 'CPI_Std', 'CPI_Var',
           'Unemployment_Mean', 'Unemployment_Median', 'Unemployment_Max', 'Unemployment_Min', 'Unemployment_Std', 'Unemployment_Var',
           'Weekly_Sales']

print("Shape of dataset:", data.shape)

data.head()
```

### 3.5 Data Preprocessing - Feature Engineering

In this phase, new features are engineered from the 'Date' column. Useful features such as year, month, day, week of the year, and quarter is extracted from the date column. Moreover, lag features are extracted from the target to help improve modeling performance.

```python
# Engineer new features from Date
# Sort Dates
data = data.sort_values('Date')

# Assuming 'Date' is a column in the train_data DataFrame
data['Date'] = pd.to_datetime(data['Date'])

# Extract useful information from the date column
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Day'] = data['Date'].dt.day
data['WeekOfYear'] = data['Date'].dt.isocalendar().week
data['Quarter'] = data['Date'].dt.quarter

# Set the "Date" column as the index
data.set_index('Date', inplace=True)
```

```python
data['Weekly_Sales_Lag1'] = data['Weekly_Sales'].shift(1)
data['Weekly_Sales_Lag2'] = data['Weekly_Sales'].shift(2)
data['Weekly_Sales_Lag3'] = data['Weekly_Sales'].shift(3)

data[['Weekly_Sales_Lag1', 'Weekly_Sales_Lag2', 'Weekly_Sales_Lag3']] = data[['Weekly_Sales_Lag1', 'Weekly_Sales_Lag2', 'Weekly_Sales_Lag3']].fillna(method='bfill')
```

```python
# Rearrange columns
data = data[['Year', 'Month', 'Day', 'WeekOfYear', 'Quarter', 
       'Holiday_Flag', 'Temperature_Mean', 'Temperature_Median',
       'Temperature_Max', 'Temperature_Min', 'Temperature_Std',
       'Temperature_Var', 'Fuel_Price_Mean', 'Fuel_Price_Median',
       'Fuel_Price_Max', 'Fuel_Price_Min', 'Fuel_Price_Std', 'Fuel_Price_Var',
       'CPI_Mean', 'CPI_Median', 'CPI_Max', 'CPI_Min', 'CPI_Std', 'CPI_Var',
       'Unemployment_Mean', 'Unemployment_Median', 'Unemployment_Max',
       'Unemployment_Min', 'Unemployment_Std', 'Unemployment_Var',
       'Weekly_Sales_Lag1', 'Weekly_Sales_Lag2', 
       'Weekly_Sales']]
```

### 3.6 EDA

EDA (Exploratory Data Analysis) involves examining and visualizing the dataset to gain insights into data.

```python
# Set the figure size and DPI
plt.figure(figsize=(16, 10), dpi=300)

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Plot the correlation matrix as a heatmap using a pleasing color scheme
plt.imshow(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar()  # Add colorbar for reference

# Customize the plot
plt.title('Correlation Matrix')
tick_labels = correlation_matrix.columns
plt.xticks(np.arange(len(tick_labels)), tick_labels, rotation=90)
plt.yticks(np.arange(len(tick_labels)), tick_labels)

# Save the plot in high quality
plt.savefig('images/correlation_matrix_plot.png', dpi=300)

# Show the plot
plt.show()
```

### 3.7 Data Splitting

In this phase, the data is split into a Training set (80%) and a Testing set (20%). This division allows for separate subsets of data to be used for model training and evaluation to assess performance.

```python
# Split the data into train and test sets
train_data, test_data = train_test_split(data, test_size=0.20, shuffle=False)

# Print the number of records and percentages for each set
print("Train set:")
print("Number of records:", train_data.shape[0])
print("Percentage: {:.0%}".format(train_data.shape[0] / data.shape[0]))

print("\nTest set:")
print("Number of records:", test_data.shape[0])
print("Percentage: {:.0%}".format(test_data.shape[0] / data.shape[0]))
```

### 3.8 Data Preprocessing - Feature Scaling

Feature scaling is an essential preprocessing step that ensures all input features are on a similar scale. By applying the StandardScaler, the dataset is normalized, enhancing the convergence and performance of neural networks, and facilitating accurate and robust predictions.

```python
# Columns to be scaled
columns_to_scale = [col for col in data.columns if col not in ['Weekly_Sales', 'Holiday_Flag']]

# Apply StdScaler scaling to features
scaler = StandardScaler()
train_data[columns_to_scale] = scaler.fit_transform(train_data[columns_to_scale])

# Apply StdScaler scaling to validation and test sets using the scaler fitted on the training data
test_data[columns_to_scale] = scaler.transform(test_data[columns_to_scale])
```

### 3.9 Model Training and Evaluation

The model training and evaluation section utilizes an LSTM-based architecture for predicting aggregated weekly sales. In the LSTM-based architecture, both LSTM neurons and dense neurons compute weighted sums of their inputs. However, the difference lies in how these weighted sums are calculated and the functionality of the neurons. LSTM neurons differ from dense neurons as they are specifically designed for handling sequential and time-dependent data, capturing long-term dependencies and temporal patterns through memory cells and gating mechanisms. The dataset was earlier split into train and test sets, with train sets further divided for fine-tuning the model. Different sequence-to-one and sequence-to-sequence models with various LSTM based architectures are explored, and evaluation is performed using MSE and RMSE metrics on the test set to measure prediction accuracy. 

```python
# Predict on the train set
trainX = []
trainY = []

# input length
n_past = 4

# output length
n_future = 1

# Iterate over the range of indices for creating train samples
for i in range(n_past, len(train_data) - n_future + 1):
    # Extract a slice of the input data for the current train sample and convert it to a numpy array
    input_slice = train_data.iloc[i - n_past:i, :-1].values  # Exclude the last column
    # Append the input slice to the trainX list
    trainX.append(input_slice)

    # Extract the target value for the current train sample and convert it to a numpy array
    target_value = train_data.iloc[i + n_future - 1:i + n_future, -1].values  # Use the last column
    # Append the target value to the trainY list
    trainY.append(target_value)

# Convert the trainX and trainY lists to numpy arrays
trainX, trainY = np.array(trainX), np.array(trainY)

# Print the shapes of the testX and testY arrays
print('trainX shape:', trainX.shape)
print('trainY shape:', trainY.shape)
```

```python
# fix random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
weight_initializer=tf.keras.initializers.GlorotUniform(seed=42)

# Set environment variables for reproducibility
os.environ['PYTHONHASHSEED'] = '42'
os.environ['TF_DETERMINISTIC_OPS'] = '1'

# Set number of threads for reproducility
tf.config.threading.set_inter_op_parallelism_threads(1)
tf.config.threading.set_intra_op_parallelism_threads(1)

# Define model
model = Sequential()
model.add(LSTM(16, activation='relu', input_shape=(trainX.shape[1], trainX.shape[2]), 
               return_sequences=True, kernel_initializer=weight_initializer))
model.add(LSTM(32, activation='relu', input_shape=(trainX.shape[1], trainX.shape[2]), return_sequences=True))
model.add(LSTM(64, activation='relu', input_shape=(trainX.shape[1], trainX.shape[2]), return_sequences=True))
model.add(LSTM(128, activation='relu', input_shape=(trainX.shape[1], trainX.shape[2]), return_sequences=True))
model.add(LSTM(128, activation='relu', input_shape=(trainX.shape[1], trainX.shape[2]), return_sequences=True))
model.add(LSTM(64, activation='relu', input_shape=(trainX.shape[1], trainX.shape[2]), return_sequences=True))
model.add(LSTM(32, activation='relu', input_shape=(trainX.shape[1], trainX.shape[2]), return_sequences=True))
model.add(LSTM(16, activation='relu', input_shape=(trainX.shape[1], trainX.shape[2]), return_sequences=False))
model.add(Dense(trainY.shape[1]))

model.summary()

# compile model
learning_rate = 0.01
optimizer = Adam(learning_rate=learning_rate)
model.compile(loss='mse', optimizer=optimizer)

# Define callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=15)

# Get validation split percentage
val_split_percentage = 0.1

# Fit the model
history = model.fit(trainX, trainY, epochs=100, batch_size=1, validation_split=val_split_percentage, shuffle=False, 
                    verbose=1, callbacks=[early_stopping])

# Plot the training and validation loss
light_red = (0.8, 0.2, 0.2)  # RGB values for a lighter red color
light_blue = (0.2, 0.2, 0.8)  # RGB values for a lighter blue color


# Train loss vs Val loss
plt.figure(figsize=(10, 3), dpi=300)
plt.plot(history.history['loss'], label='Training Loss', color= light_red)
plt.plot(history.history['val_loss'], label='Validation Loss', color= light_blue)
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.tight_layout() 
plt.show()
```

```python
# Calculate the index to split the data
split_index = int(len(trainX) * (1 - val_split_percentage))

# Predict on the training set
train_predictions = model.predict(trainX[:split_index])

# Calculate MAE on the training set
train_mse = mean_squared_error(trainY[:split_index], train_predictions)
# Calculate RMSE on the training set
train_rmse = np.sqrt(train_mse)
# print results
print("MSE on training set:", round(train_mse, 4))
print("RMSE on training set:", round(train_rmse, 4))
print(" ")

# Predict on the validation set
val_predictions = model.predict(trainX[split_index-1:])

# Calculate MAE on the validation set
val_mse = mean_squared_error(trainY[split_index-1:], val_predictions)
# Calculate RMSE on the validation set
val_rmse = np.sqrt(val_mse)
# print results
print("MSE on validation set:", round(val_mse, 4))
print("RMSE on validation set:", round(val_rmse, 4))
```

```python
# Plot actual and predicted values
plt.figure(figsize=(10, 3), dpi=300)
plt.plot(trainY, label='Actual', color='grey')
plt.plot(range(split_index), train_predictions, label='Predicted (Training)', color=light_red)
plt.plot(range(split_index-1, len(trainY)), val_predictions, label='Predicted (Validation)', color=light_blue)
plt.title('Actual vs. Predicted Target Values')
plt.xlabel('Time')
plt.ylabel('Target Value')
plt.legend()
plt.legend(loc='upper left')  # Set the legend position to the top left
plt.tight_layout()
plt.show()
```

```python
# Predict on the test set
testX = []
testY = []

# Iterate over the range of indices for creating test samples
for i in range(n_past, len(test_data) - n_future + 1):
    # Extract a slice of the input data for the current test sample and convert it to a numpy array
    input_slice = test_data.iloc[i - n_past:i, :-1].values  # Exclude the last column
    # Append the input slice to the testX list
    testX.append(input_slice)

    # Extract the target value for the current test sample and convert it to a numpy array
    target_value = test_data.iloc[i + n_future - 1:i + n_future, -1].values  # Use the last column
    # Append the target value to the testY list
    testY.append(target_value)

# Convert the testX and testY lists to numpy arrays
testX, testY = np.array(testX), np.array(testY)

# Print the shapes of the testX and testY arrays
print('testX shape:', testX.shape)
print('testY shape:', testY.shape)

# Predict on the test set
test_predictions = model.predict(testX)

# Calculate MAE on the test set
test_mse = mean_squared_error(testY, test_predictions)
# Calculate RMSE on the training set
test_rmse = np.sqrt(test_mse)
# print results
print("MSE on test set:", round(test_mse, 4))
print("RMSE on test set:", round(test_rmse, 4))
```

```python
# Plot actual and predicted values
plt.figure(figsize=(10, 3), dpi=300)
plt.plot(testY, label='Actual', color = light_red)
plt.plot(test_predictions, label='Predicted', color = light_blue)
plt.title('Actual vs. Predicted Target Values')
plt.xlabel('Time')
plt.ylabel('Target Value')
plt.legend()
plt.tight_layout() 
plt.legend(loc='upper left')  # Set the legend position to the top left
plt.show()
```
